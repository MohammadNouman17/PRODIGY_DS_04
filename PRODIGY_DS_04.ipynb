```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Task4DS.csv")
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.describe
```

```python
Sentiments_sum = df['Sentiment'].value_counts()
Sentiments_sum
```

```python
plt.figure(figsize=(6, 3))
Sentiment_sum.plot(kind='bar', color=['red', 'green', 'skyblue', 'orange'])
plt.title('Distribution of Twitter Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
Entity_data = df[df['Entity'].str.contains('LeagueOfLegends', case=False)]
Entity_Sentiment_Sum = Entity_data['Sentiment'].value_counts()
Entity_Sentiment_Sum
```

```python
plt.figure(figsize=(6, 6))
plt.pie(Entity_Sentiment_Sum, labels=Entity_Sentiment_Sum.index, autopct='%1.1f%%', startangle=0)
plt.title('Distribution of Sentiment for the Entity "LeagueOfLegends"')
plt.show()
```
